```python
import pandas as pd

# Load dataset with correct encoding
df = pd.read_csv(r"C:\Users\Dell\Desktop\Superstore.csv", encoding="latin-1")

# Show first 5 rows
df.head()
```

```python
# Shape of dataset
print("Rows & Columns:", df.shape)

# Column names & data types
print(df.info())

# Basic statistics
print(df.describe())

# Missing values
print(df.isnull().sum())
```

```python
# Convert Order Date & Ship Date to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Check unique values in Category/Segment
print(df['Category'].unique())
print(df['Segment'].unique())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

category_sales = df.groupby("Category")["Sales"].sum().sort_values()
sns.barplot(x=category_sales.values, y=category_sales.index)
plt.title("Total Sales by Category")
plt.show()
```

```python
region_sales = df.groupby("Region")["Sales"].sum().sort_values()
sns.barplot(x=region_sales.values, y=region_sales.index)
plt.title("Sales by Region")
plt.show()
```

```python
sns.scatterplot(data=df, x="Sales", y="Profit", hue="Category")
plt.title("Profit vs Sales")
plt.show()
```


1. Total Sales by Category

Technology contributes the highest sales among all categories.

Office Supplies has the lowest sales, though it might be purchased frequently in small amounts.

Furniture sales are significant but not as strong as Technology.

2. Total Sales by Region

The West region leads in overall sales, showing strong market presence.

The East region comes second, also showing high demand.

Central and South regions underperform, indicating potential areas for market growth.

3. Profit vs Sales (Scatter Plot)

Higher sales don’t always result in higher profits.

Technology products generally provide both high sales and higher profits.

Furniture has cases where large sales lead to negative profit, possibly due to heavy discounts or high costs.

Office Supplies usually stay in a low sales–low profit range.

 Conclusion

- Technology is the most profitable category.  
- West and East are the best-performing regions.  
- Furniture has high sales but low profit margins, needing optimization.
